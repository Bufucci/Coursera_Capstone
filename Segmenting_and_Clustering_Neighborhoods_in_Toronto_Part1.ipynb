```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

print("Libraries Imported !")
```

### *1. Scraping the List of postal codes of Canada :*


```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text
soup = BeautifulSoup(source, 'xml')
table = soup.find('table')
table
```

#### *Dataframe will consist of three columns : PostalCode, Borough, and Neighborhood*

```python
column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
df
```

#### *Search all the postcode, borough, neighborhood :*

```python
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data

df.head()
```

#### *Data Cleaning :*

Removing rows where Borough is 'Not assigned'

```python
df_Cleaned = df[df['Borough']!='Not assigned'].copy()
df_Cleaned.set_index('Postalcode', inplace=True)
df_Cleaned.reset_index(inplace=True)
df_Cleaned.head()
```

Replacing Neighborhood 'Not assigned' by the associated Borough 

```python
df_Cleaned['Neighborhood'][df_Cleaned['Neighborhood']=='Not assigned']= df_Cleaned['Borough'][df_Cleaned['Neighborhood']=='Not assigned']
df_Cleaned.head(10)

#df_check=df_Cleaned[['Postalcode']].count(axis=1).sum()
#df_check
```
